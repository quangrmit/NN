```python
import pandas as pd
from linear_regression import fit, predict
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./X_train.csv')
labels = pd.read_csv('./y_train.csv')
```

```python
# Generate a dataset

def gen_data():
    dataf = []
    for i in range(100):
        row = []
        for j in range(3):
            row.append(random.randrange(0, 6))
        dataf.append(row)

    print(dataf)
    return dataf

def gen_labels():
    labs = [random.randrange(10, 20) for _ in range(100)]
    return labs


X = pd.DataFrame(gen_data())
y = pd.Series(gen_labels())
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
model.score(X, y)
```

```python
params = fit(X, y)
```

```python
y_pred_mine = predict(X, params)
```

```python
y_pred_model = model.predict(X)
mean_absolute_error(y_pred_model, y)
```
